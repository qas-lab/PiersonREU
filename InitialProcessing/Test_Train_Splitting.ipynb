```python
import pandas as pd
import os
```

```python
fulldata = pd.read_csv('/home/rpierson/PiersonREU/extracted/eclipse_platform.csv')
fulldata
```

```python
import os

fulldata = fulldata.sort_values(by='Created_time')
fulldata.head()
```

```python
n = len(fulldata)
split_index = int(.8 * n)
train_dataset = fulldata.iloc[:split_index]
```

```python
test_dataset = fulldata.drop(train_dataset.index)

extract_dir = '/home/rpierson/PiersonREU/extracted'

```

```python
train_features = train_dataset.copy()
test_features = test_dataset.copy()

train_dataset = train_dataset[['Issue_id','Priority','Title', 'Description', 'Component']]
test_dataset = test_dataset[['Issue_id','Priority','Title', 'Description', 'Component']]

train_priority_labels = train_features.pop('Priority')
test_priority_labels = test_features.pop('Priority')

train_priority_path = os.path.join(extract_dir, 'train_priority.csv')
test_priority_path = os.path.join(extract_dir, 'test_priority.csv')

train_priority_labels.to_csv(train_priority_path, index = False)
test_priority_labels.to_csv(test_priority_path, index = False)
```

```python
train_file = os.path.join(extract_dir, 'train_dataset.csv')
test_file = os.path.join(extract_dir, 'test_dataset.csv')

train_dataset.to_csv(train_file, index = False)
test_dataset.to_csv(test_file, index = False)
```

```python
extract_dir = '/home/rpierson/PiersonREU/extracted'
training_file = os.path.join(extract_dir, 'train_dataset_notpreprocessed.csv')
testing_file = os.path.join(extract_dir, 'test_dataset_notpreprocessed.csv')

train_dataset_notpreprocessed = train_dataset.copy()
test_dataset_notpreprocessed = test_dataset.copy()

train_dataset_notpreprocessed.to_csv(training_file, index = False)
test_dataset_notpreprocessed.to_csv(testing_file, index = False)
```

```python
train_dataset_notpreprocessed['Combined_Text'] = train_dataset_notpreprocessed["Title"] + " " + train_dataset_notpreprocessed["Component"] + " " + train_dataset_notpreprocessed["Description"]
subset_size = train_dataset_notpreprocessed.size/10
for x in range(10):
    start_index = int(x * subset_size)
    end_index = int((x + 1) * subset_size)
    subset = train_dataset_notpreprocessed.iloc[start_index:end_index]
    training_file = os.path.join(extract_dir, f'train_dataset_{x}.csv')
    subset.to_csv(training_file, index=False)
```
